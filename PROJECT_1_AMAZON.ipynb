# **1. Data Cleaning**

```python
df = pd.read_csv("/content/drive/MyDrive/data/amazon.csv")
df.info()
df.isna().sum()
df
```

```python
df.dropna(subset = ['rating_count'], inplace = True)
df.info()
df.drop_duplicates(inplace = True)
```

```python
df.describe()
df.tail()
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = df['rating_count'].replace({',': ''}, regex = True).astype(float)

df['discounted_price'] = df['discounted_price'].replace({'₹': ''}, regex = True)
df['discounted_price'] = df['discounted_price'].replace({',': ''}, regex = True).astype(float)

df['actual_price'] = df['actual_price'].replace({'₹': ''}, regex = True)
df['actual_price'] = df['actual_price'].replace({',': ''}, regex = True).astype(float)

df['rating_count'] = pd.to_numeric(df['rating_count'], errors = 'coerce')

df['discount_percentage'] = df['discount_percentage'].replace({'%': ''}, regex = True).astype(float)

df.dropna(subset = ['rating'], inplace = True)
df = df[df['discount_percentage'] > 0]
df['category'].value_counts()
df.info()
```

```python
classified_categories = []

for category in df["category"]:
    main_category = category.split("|")[0]
    classified_categories.append(main_category)

df["Classified Category"] = classified_categories
df['Classified Category'] = df['Classified Category'].replace(
    ['MusicalInstruments', 'HomeImprovement', 'Car&Motorbike', 'Health&PersonalCare'], 'Others')
df['Classified Category'].value_counts()
```

```python
def categorise_discount(discount):
    if 0 <= discount <= 20:
        return "0 - 20"
    elif 20 < discount <= 40:
        return "20 - 40"
    elif 40 < discount <= 60:
        return "40 - 60"
    elif 60 < discount <= 80:
        return "60 - 80"
    elif discount > 80:
        return "> 80"
    else:
      return "Invalid"

df['Discount Categories'] = df['discount_percentage'].apply(categorise_discount)
df_new = df.groupby('Discount Categories').agg(
    mean_discount_percentage = ('discount_percentage', 'mean'),
    sum_discount_percentage = ('discount_percentage', 'sum'),
    mean_rating = ('rating', 'mean'),
    sum_rating = ('rating', 'sum'),
    sum_rating_count = ('rating_count', 'sum'),
    mean_rating_count = ('rating_count', 'mean'),
    count=('discount_percentage', 'count')
)
df_new.reset_index(inplace = True)
df_new
```

```python
import pandas as pd

import nltk

from nltk.sentiment.vader import SentimentIntensityAnalyzer

from nltk.corpus import stopwords

from nltk.tokenize import word_tokenize

from nltk.stem import WordNetLemmatizer

# download nltk corpus (first time only)
import nltk

nltk.download('all')
# create preprocess_text function
def preprocess_text(text):

    # Tokenize the text
    tokens = word_tokenize(text.lower())
    # Remove stop words
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]
    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    # Join the tokens back into a string
    processed_text = ' '.join(lemmatized_tokens)

    return processed_text
# apply the function df
df['review_title'] = df['review_title'].apply(preprocess_text)
```

```python
# initialize NLTK sentiment analyzer
analyzer = SentimentIntensityAnalyzer()
# create get_sentiment function
def get_sentiment(text):
    scores = analyzer.polarity_scores(text)
    print(scores['pos'])
    sentiment = 1 if scores['pos'] > 0.05 else 0
    return sentiment
# apply get_sentiment function

df['sentiment'] = df['review_content'].apply(get_sentiment)

df.tail(10)
```

```python
from sklearn.metrics import confusion_matrix
df['Positive'] = df['sentiment']

print(confusion_matrix(df['Positive'], df['sentiment']))
from sklearn.metrics import classification_report

print(classification_report(df['Positive'], df['sentiment']))
```
